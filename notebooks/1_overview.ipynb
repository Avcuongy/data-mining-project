#  Univariate Analysis

## Category

```python
import matplotlib.pyplot as plt

data = products["product_category_name"].value_counts().sort_values()

plt.figure(figsize=(12, 15))
ax = data.plot(kind="barh")

for i, v in enumerate(data):
    ax.text(v, i, str(v), va='center')

plt.xlabel("Number of Products")
plt.ylabel("Product Category")
plt.title("Product Category")
plt.show()
```

## Numerical

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols = [
    "product_name_lenght",
    "product_description_lenght",
    "product_photos_qty",
    "product_weight_g",
    "product_length_cm",
    "product_height_cm",
    "product_width_cm"
]

fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(14, 18))
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.histplot(products[col], kde=True, ax=axes[i], bins=30)
    axes[i].set_title(f"{col}")

# Xóa ô thừa
fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(18, 20))
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.boxplot(x=products[col], ax=axes[i])
    axes[i].set_title(f"Boxplot of {col}")

# Xóa ô thừa
fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()
```
